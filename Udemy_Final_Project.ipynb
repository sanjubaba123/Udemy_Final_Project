```python
df=pd.read_csv('mammographic_masses.data.csv',na_values=['?'])
df.head()
```

```python
df.describe()
```

```python
#it will find all the rows which has atleast one NaN data or missing data
df.loc[(df['age'].isnull())|
       (df['shape'].isnull())|
       (df['margin'].isnull())|
       (df['density'].isnull())
      ]
```

```python
df.head()
```

```python
final_df=df.dropna()
final_df.head()
```

```python
final_df.describe()
```

```python
x=final_df.drop(['BI_READS'],axis='columns')
x.head()
```

```python
y=final_df[['BI_READS']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.head()
```

```python
len(x_train)
```

```python
x_test.head()
```

```python
len(x_test)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from pandas import DataFrame
y_predicted=pd.DataFrame(model.predict(x_test))
y_predicted.head()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
sb.heatmap(cm,annot=True)
plt.xlabel('y_test',fontsize=30)
plt.ylabel('y_predicted',fontsize=30)
plt.show()
```

```python
dt=tree.DecisionTreeClassifier(min_samples_split=10)
```

```python
dt.fit(x_train,y_train)
dt.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(tree.DecisionTreeClassifier(),x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier 
model2=RandomForestClassifier(n_estimators=10)
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
model3=SVC()
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
#it has more accuracy than decision tree
```

```python
#for finding the value of k we will use elbow technique
from sklearn.cluster import KMeans
k_range=range(1,10)
sse=[]
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
sse
    
```

```python
plt.xlabel('K',fontsize=30)
plt.ylabel('sum of squared error',fontsize=30)
plt.plot(k_range,sse)
plt.show()
```
